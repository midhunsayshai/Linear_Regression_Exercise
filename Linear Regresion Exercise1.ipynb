### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

### Read the data

```python
Canada=pd.read_csv('canada_per_capita_income.csv')
Canada
```

```python
Canada.columns
```

```python
Canada.rename(columns={'per capita income (US$)':'percapita'},inplace='True') ##Renamed the column name to percapita
Canada.percapita
```

```python
plt.xlabel('Percapita')
plt.ylabel('Year')
plt.scatter(Canada.percapita,Canada.year,color='red',marker="*")
```

### To find the independent features/x

```python
x=Canada.drop('percapita',axis=1)
x
```

```python
y=Canada.percapita
y
```

### Importing Sklearn for importing linear model

```python
from sklearn import linear_model
```

### Create linear regression object

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(x,y)
```

```python
Predict_percapita=reg.predict([[2020]])
Predict_percapita
```
